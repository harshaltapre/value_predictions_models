```python
import numpy  as np
import pandas as pd
import sklearn.datasets
```

```python
# import LogisticRegression
from sklearn.linear_model import LogisticRegression
```

```python
# accuracy_score
from sklearn.metrics import accuracy_score
```

```python
# importing data set
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# printing the dataset
breast_cancer_dataset
```

```python
#load the data set into data frame
df=pd.DataFrame(breast_cancer_dataset.data ,columns=breast_cancer_dataset.feature_names)
```

```python
# print the first five rows
df.head()
```

```python
#adding the target(Label) col in df
df['Label'] = breast_cancer_dataset.target
```

```python
# print the last 2 rows
df.tail()
```

```python
#no. of col and rows
df.shape
```

```python
# informmation of dataset
df.info()
```

```python
# find the null values and sum
df.isnull().sum()
```

```python
#Statistical mesure of data
df.describe()
```

```python
# Checking the distrubution of target varible
df['Label'].value_counts()
# 1 is Benign
# 0 is Milignamt
```

```python
# Groupby lable col and fin the mean for all the col
df.groupby('Label').mean()
```

```python
df
```

```python
# Input Feature -X (All col except label)
# target COl-Y  (label col)
```

```python
#seprating Features and target colums
X = df.drop(columns= 'Label',axis =1)
Y = df['Label']
```

```python
#printing the feature

print(X)
```

```python
#print target varible
print (Y)
```

```python
from sklearn.model_selection import train_test_split
#Splitting the data into training data and testing data
X_train , X_test ,Y_train , Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
df
```

```python
#Model Training
#Logistic Regression is most useful for binary value classifictaion.
model = LogisticRegression()
#Training the ML model using training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#Checking Accuracy Score of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy on training data =", training_data_accuracy)
```

```python
#Checking Accuracy Score of testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy on testing data =", testing_data_accuracy)
```
