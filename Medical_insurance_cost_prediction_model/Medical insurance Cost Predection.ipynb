Data Preprocessing

Encoding Catogerical

```python
# Encoding the sex column
insurance_dataset.replace({'sex':{'male' : 1, 'female' : 0}}, inplace = True)

# Encoding Smoker Column
insurance_dataset.replace({'smoker':{'yes' : 1, 'no' : 0}}, inplace = True)

# Encoding Region Column
insurance_dataset.replace({'region':{'southeast' : 0, 'southwest' : 1, 'northeast': 2, 'northwest' : 3}}, inplace = True)
```

```python
insurance_dataset
```

```python
X = insurance_dataset.drop(columns = 'charges', axis = 1)
```

```python
Y = insurance_dataset.charges
```

Splliting the data  into traning and testing data

```python
X_train , X_test, Y_train, Y_test = train_test_split(X , Y, test_size = 0.2, random_state = 2)
```

```python
print( X.shape ,X_train.shape,X_test.shape)
```

Modal **Tranning**

```python
# Calling Linear Regression Model
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

Model evaluaion

```python
# predection on training the Data
training_data_predection = regressor.predict(X_train)
```

```python
# R Square Value
r2_train = metrics.r2_score(Y_train,training_data_predection)
print( "R Square Value " , r2_train )
```

```python
# R square value on testing data

testing_data_prediction = regressor.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test,testing_data_prediction)
print( "R Square Value " , r2_test )
```
