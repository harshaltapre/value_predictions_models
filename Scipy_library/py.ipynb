```python
import pandas as pd
from scipy.stats import skew, kurtosis
from pandas import plotting

column_name = 'FSIQ'
skewness = skew(df[column_name])
kurt = kurtosis(df[column_name])
print(skewness)
print(kurt)

# Displaying Relationship Between Different Parameters
df['Height'].fillna(method='pad', inplace=True)
plotting.scatter_matrix(df[['VIQ', 'MRI_Count', 'Height']], c=(df['Gender'] == 'Female'), marker='o', alpha=1, cmap='winter')
fig = plt.gcf()
fig.suptitle("blue:male,green:female", size=13)
plt.show()
```
