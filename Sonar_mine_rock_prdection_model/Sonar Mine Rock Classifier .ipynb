```python
import numpy  as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Importing LogisticRegression Model
from sklearn.linear_model import LogisticRegression
```

```python
#Importing accuracy score function
from sklearn.metrics import accuracy_score
```

```python

#loading the data from csv file to a pandas dataframe
sonar_dataset = pd.read_csv('sonar data.csv',header=None)
```

```python
X=sonar_dataset.iloc[:,0:60]
Y=sonar_dataset.iloc[:,60]
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(X , Y, test_size = 0.2, random_state = 2)
```

```python
print( X.shape ,X_train.shape,X_test.shape)
```

```python
#Model Training
#Logistic Regression is most useful for binary value classifictaion.
model = LogisticRegression()
#Training the ML model using training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#Checking Accuracy Score of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy on traning  data =", training_data_accuracy)
```

```python
#Checking Accuracy Score of testing data
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy on testing  data =", training_data_accuracy)
```
